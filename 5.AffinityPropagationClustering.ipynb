```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn.metrics import silhouette_score
list1 = []
for i in range(-900, -9900, -100):
    res =AffinityPropagation(preference=i) 
    s_score=silhouette_score(X,res.fit_predict(X), metric='euclidean')
    list1.append(s_score)
plt.plot(range(-900, -9900, -100), list1)
plt.title('The silhouette_score Method')
plt.xlabel('Number of bandwidth')
plt.ylabel('Silhouette_score')
plt.show()
```

```python
from sklearn.cluster import AffinityPropagation
model =AffinityPropagation(preference=-9800) 
res=model.fit_predict(X)
res
```

```python
supervised=dataset
supervised['Cluster_group']=res
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```

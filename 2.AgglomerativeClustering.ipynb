```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
X
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
list1 = []
for i in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters = i)
    s_score=silhouette_score(X,agglo.fit_predict(X), metric='euclidean')
    list1.append(s_score)
plt.plot(range(2, 11), list1)
plt.title('The silhouette_score Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette_score')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clusmodel = AgglomerativeClustering(n_clusters = 5)
label = clusmodel.fit_predict(X)
```

```python
supervised=dataset
supervised['Cluster_group']=label
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```

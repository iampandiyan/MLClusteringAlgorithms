```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:, 3:5].values
X
```

```python
from sklearn.cluster import BisectingKMeans
list1 = []
for i in range(1, 11):
    bkmeans =BisectingKMeans(n_clusters=i, init = 'k-means++',random_state=42)
    bkmeans.fit(X)
    list1.append(bkmeans.inertia_)
plt.plot(range(1, 11), list1)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
list1 = []
for i in range(2, 11):
    bisectKmeans = BisectingKMeans(n_clusters=i, init = 'k-means++',random_state=0)
    s_score=silhouette_score(X,bisectKmeans.fit_predict(X), metric='euclidean')
    list1.append(s_score)
plt.plot(range(2, 11), list1)
plt.title('The silhouette_score Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette_score')
plt.show()
```

```python
from sklearn.cluster import BisectingKMeans
bisectKmeans = BisectingKMeans(n_clusters=6, init = 'k-means++',random_state=0)
y_bkmeans=bisectKmeans.fit_predict(X)
y_bkmeans
```

```python
supervised=dataset
supervised['Cluster_group']=y_bkmeans
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
